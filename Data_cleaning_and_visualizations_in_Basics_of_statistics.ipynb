```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/sales_data_with_discounts.csv')

print(df.head())

```

```python
numerical_columns = df.select_dtypes(include=[float, int]).columns.tolist()

statistics = pd.DataFrame(columns=['mean', 'median', 'mode', 'std'])

for column in numerical_columns:
    statistics.loc[column, 'mean'] = df[column].mean()
    statistics.loc[column, 'median'] = df[column].median()
    statistics.loc[column, 'mode'] = df[column].mode()[0]
    statistics.loc[column, 'std'] = df[column].std()

print("Descriptive Statistics for Numerical Columns:")
print(statistics)

for column in numerical_columns:
    print(f"\nColumn: {column}")
    print(f"Mean: {statistics.loc[column, 'mean']}")
    print(f"Median: {statistics.loc[column, 'median']}")
    print(f"Mode: {statistics.loc[column, 'mode']}")
    print(f"Standard Deviation: {statistics.loc[column, 'std']}")
    print(f"Interpretation: The mean provides the average value. The median gives the middle value when sorted, which is useful for skewed distributions. The mode shows the most frequent value, and the standard deviation indicates the spread or variability of the data.")
```

```python
numerical_columns = df.select_dtypes(include=[float, int]).columns.tolist()

for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
    print(f"Histogram Analysis for {column}:")
    print(f"Skewness: {df[column].skew()}")
    print(f"Presence of outliers: {df[column].describe(percentiles=[.01, .99])}\n")

for column in numerical_columns:
    plt.figure(figsize=(15, 3))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()

    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    print(f"Boxplot Analysis for {column}:")
    print(f"Interquartile Range (IQR): {iqr}")
    print(f"Lower Bound: {lower_bound}")
    print(f"Upper Bound: {upper_bound}")
    print(f"Number of outliers: {outliers.count()}\n")
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

for column in categorical_columns:
    plt.figure(figsize=(15, 3))
    df[column].value_counts().plot(kind='bar')
    plt.title(f'Bar Chart of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

    print(f"Bar Chart Analysis for {column}:")
    value_counts = df[column].value_counts()
    print(value_counts)
    print(f"Most frequent category: {value_counts.idxmax()} with {value_counts.max()} occurrences")
    print(f"Least frequent category: {value_counts.idxmin()} with {value_counts.min()} occurrences\n")
```

```python
numerical_columns = df.select_dtypes(include=[float, int]).columns.tolist()

standardized_df = df.copy()
for column in numerical_columns:
    mean = df[column].mean()
    std = df[column].std()
    standardized_df[column] = (df[column] - mean) / std

for column in numerical_columns:
    plt.figure(figsize=(15, 3))

    plt.subplot(1, 2, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Original {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    plt.subplot(1, 2, 2)
    sns.histplot(standardized_df[column], kde=True)
    plt.title(f'Standardized {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    plt.show()

    print(f"Statistics for {column} before standardization:")
    print(f"Mean: {df[column].mean()}, Standard Deviation: {df[column].std()}")
    print(f"Statistics for {column} after standardization:")
    print(f"Mean: {standardized_df[column].mean()}, Standard Deviation: {standardized_df[column].std()}\n")
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

encoded_df = pd.get_dummies(df, columns=categorical_columns)

print("Dataset after one-hot encoding:")
print(encoded_df.head())
print("\nNew columns created after one-hot encoding:")
print(encoded_df.columns.tolist())
```

```python
#Conclusion

##Key Findings from Descriptive Analytics and Data Visualizations:

1. Descriptive Statistics:
   - The mean, median, mode, and standard deviation gave us insights into the data's central tendency and spread.
   - High standard deviations indicated more variability, while low standard deviations suggested consistency.

2. Histograms and Boxplots:
   - Histograms showed the distribution of numerical variables, revealing skewness and outliers.
   - Boxplots highlighted outliers and the spread of data, showing extreme values and unusual distributions.

3. Bar Charts for Categorical Columns:
   - Bar charts displayed the frequency of categories, revealing imbalances in the data.

##Importance of Data Preprocessing:

1. Standardization:
   - Standardizing numerical variables ensures uniformity, helping models that rely on distance calculations.

2. One-Hot Encoding:
   - One-hot encoding transforms categorical variables into a usable format for machine learning algorithms without implying any order.

Preprocessing steps like standardization and one-hot encoding are crucial for preparing data, improving model performance, and ensuring reliable results.
```
